```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ml2_train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
obj_col = ['State']
df = pd.get_dummies(df, columns=obj_col)
```

```python
df.drop(['Serial Number','Main Office','Branch Number','County','City','Established Date','Acquired Date'],axis=1,inplace=True)
```

```python
df.info()
```

```python
X=df.drop('2016 Deposits',axis=1)
```

```python
y= df['2016 Deposits']
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
datadm = xgb.DMatrix(data=X,label=y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
xg_reg = xgb.XGBRegressor(objective ='reg:linear')
```

```python
xg_reg.fit(X_train,y_train)
preds = xg_reg.predict(X_test)
```

```python
preds
```

```python
y_test
```

```python
for i in range(len(preds)):
    if preds[i] == y_test[i]:
        print('Yes')
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```
