```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Yes_Bank_Train.csv')
```

```python
df =df.drop(['serial number'],axis=1)
X = df.drop(['credit_amount'],axis=1)
y = df['credit_amount']
obj_col=X.select_dtypes(include='object')
obj_col = obj_col.columns.values
X = pd.get_dummies(X, columns=obj_col)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(X_train,label=y_train)
dtest = xgb.DMatrix(X_test,label=y_test)
```

```python
params={
    'objective':'reg:tweedie',
    'eval_metric':'rmse',
    'booster':'dart',
    'learning_rate':0.13,
    'max_depth': 2,
    'colsample_bytree': 0.5
}
evallist = [(dtest, 'eval'), (dtrain, 'train')]
```

```python
num_round=10000
bst = xgb.train(params,dtrain,num_round,evallist,early_stopping_rounds=50)
```

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
params = {
     'objective':['reg:tweedie'],
    'booster':['gblinear'],
    'learning_rate':np.arange(0.01,0.15,0.02),
    'max_depth':np.arange(1,10,1),
    'colsample_bytree':np.arange(0.1,1.1,0.1)
}
xg_reg = xgb.XGBRegressor(eval_metric='rmse',num_round=10000,early_stopping_rounds=50)
gridser = GridSearchCV(xg_reg,params,scoring='neg_mean_squared_error',n_jobs=-1)
gridser.fit(X,y)
```

```python
gridser.best_params_
```

```python
np.sqrt(-1*gridser.best_score_)
```

SUBMISSION

```python
dtrain = xgb.DMatrix(X,label=y)
params={
    'objective':'reg:tweedie',
    'eval_metric':'rmse',
    'booster':'gblinear',
    'learning_rate':0.13,
    'max_depth': 2,
    'colsample_bytree': 0.5
}
num_round=2000
bst = xgb.train(params,dtrain,num_round)
```

```python
df2 = pd.read_csv('Yes_Bank_Test_int.csv')
df2 =df2.drop(['serial number'],axis=1)
X_t = df2
obj_col=X_t.select_dtypes(include='object')
obj_col = obj_col.columns.values
X_t = pd.get_dummies(X_t, columns=obj_col)
dt = xgb.DMatrix(X_t)
```

```python
finalpred = bst.predict(dt)
```

```python
_ = pd.read_csv('Yes_Bank_Test_int.csv')
serial = _['serial number']
data = { 'serial number': serial, 'credit_amount': finalpred}
submission = pd.DataFrame(data)
submission.head()
```
